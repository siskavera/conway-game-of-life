```python
import time

import matplotlib.pyplot as plt
import numpy
import seaborn
from IPython.display import clear_output

from src.conway import get_next_state

%matplotlib inline
```

```python
def live_plot(state, figsize=(7,5), title=''):
    clear_output(wait=True)
    plt.figure(figsize=figsize)
    
    seaborn.heatmap(state, vmin=0, vmax=1)
    plt.title(title)
    plt.
    
    plt.show();
```

```python
n=20
state = numpy.random.randint(2, size=(n,n))

for i in range(10):
    state = get_next_state(state)
    live_plot(state, title=i)
    time.sleep(0.2)
```
